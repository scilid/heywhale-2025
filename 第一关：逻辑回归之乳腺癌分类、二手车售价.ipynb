#### å¼•å…¥ä¾èµ–

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_squared_error
from sklearn.preprocessing import LabelEncoder
````

#### åŠ è½½æ•°æ®åŠåŸºç¡€åˆ†æ

````python
# è·å–æ•°æ®

cancer = datasets.load_breast_cancer()
x = pd.DataFrame(cancer.data)
y = pd.DataFrame(cancer.target)
x.head(), y.head()
````

````python
x[x.index == 411]
````

````python
# æ•°æ®åŸºç¡€åˆ†æ

x.info(), y.info()
````

åœ¨å¯¹æ•°æ®ç»è¿‡åˆæ­¥åˆ†æä¹‹åï¼Œå‘ç°æ•°æ®æœªå‡ºç°ç¼ºå¤±å€¼ï¼ˆnanå€¼æˆ–ç©ºå­—ç¬¦ä¸²ç­‰ï¼‰ï¼Œä¹Ÿä¸å­˜åœ¨å¼‚å¸¸å€¼ï¼ˆæ˜æ˜¾è¶…å‡ºæ•°æ®èŒƒå›´æˆ–ä¸æ•°æ®å®šä¹‰å–å€¼åŒºé—´ä¸ç¬¦ç­‰ï¼‰ã€‚

#### è®­ç»ƒæ¨¡å‹å¹¶è®¡ç®—æŒ‡æ ‡

````python
# æ¨¡å‹è®­ç»ƒ

# 1. æ‹†åˆ†æ•°æ®è®­ç»ƒé›†å’Œæµ‹è¯•é›†ï¼Œæ¯”ä¾‹ 7:3
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
x_train.head(), x_test.head(), y_train.head(), y_test.head()
````

````python
# 2. æ„å»ºæ¨¡å‹

model = LinearRegression()
````

````python
# 3. è®­ç»ƒæ¨¡å‹

model.fit(x_train, y_train)
````

````python
# 4. ä½¿ç”¨æµ‹è¯•é›†æµ‹è¯•æ¨¡å‹è·å–æµ‹è¯•ç»“æœ

y_pred = model.predict(x_test)
y_pred[:10]
# æ­¤è¾“å‡ºçš„æ˜¯é¢„æµ‹ä¸º1 çš„æ¦‚ç‡
````

````python
# 5. è®¡ç®—æŒ‡æ ‡

# mse å‡æ–¹è¯¯å·®
# rmse å‡æ–¹æ ¹è¯¯å·®
# å¯¹äºè¿™ä¸¤ä¸ªæŒ‡æ ‡ï¼Œæ•°å€¼è¶Šå°è¶Šå¥½ï¼Œå› ä¸ºå®ƒä»¬è¡¡é‡äº†æ¨¡å‹é¢„æµ‹å€¼ä¸çœŸå®å€¼ä¹‹é—´çš„å·®å¼‚ï¼Œè¶Šå°è¡¨ç¤ºæ¨¡å‹çš„é¢„æµ‹è¶Šæ¥è¿‘çœŸå®å€¼
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mse, rmse
````

````python
# è¾“å‡ºæ¨¡å‹çš„æ–œç‡ã€å€¾æ–œé‡å’Œæµ‹è¯•é›†çš„å‡†ç¡®ç‡

model.coef_, model.intercept_, model.score(x_test, y_test)
````
