#### å¼•å…¥ä¾èµ–

```python
import pandas as pd
import numpy as np

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.metrics import accuracy_score, r2_score, mean_squared_error
```

#### åŠ è½½æ•°æ®

```python
# 1. åŠ è½½æ•°æ®

iris = load_iris()
x, y = pd.DataFrame(iris.data), iris.target
x.head(), y
```

#### è®­ç»ƒæ¨¡å‹å’Œè®¡ç®—æµ‹è¯•é›†æŒ‡æ ‡

```python
# 2. åˆ‡åˆ†æ•°æ®é›†

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
x_train, x_test, y_train, y_test
```

```python
# 3. æ„å»ºå†³ç­–æ ‘æ¨¡å‹å¹¶è®­ç»ƒæ¨¡å‹

clf = DecisionTreeClassifier(criterion='gini')

clf.fit(x_train, y_train)
```

```python
# 4. é¢„æµ‹æµ‹è¯•é›†

y_pred = clf.predict(x_test)
```

```python
# 5. è®¡ç®—æµ‹è¯•é›†çš„å‡†ç¡®ç‡

acc = accuracy_score(y_test, y_pred)
acc
```

#### åŠ è½½æ•°æ®

```python
# 1. åŠ è½½æ•°æ®

boston = pd.read_csv('/home/mw/input/boston_housing/housing.csv')
boston.head()
```

#### é¢„å¤„ç†æ•°æ®

```python
# 2. è·å–ç‰¹å¾é›†å’Œæˆ¿ä»·
x = boston.drop(['MEDV'], axis=1)
y = boston['MEDV']
x.head(), y.head()
```

```python
# 3. æµ‹è¯•é›†ä¸è®­ç»ƒé›† 7:3

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
x_train.head(), x_test.head(), y_train.head(), y_test.head()
```

#### è®­ç»ƒå›å½’æ¨¡å‹

```python
# 4. åˆ›å»º CART å›å½’æ ‘

dtr = DecisionTreeRegressor()
```

```python
# 5. è®­ç»ƒæ„é€  CART å›å½’æ ‘

dtr.fit(x_train, y_train)
```

```python
# 6. é¢„æµ‹æµ‹è¯•é›†ä¸­çš„æˆ¿ä»·

y_pred = dtr.predict(x_test)
y_pred
```

#### è®¡ç®—æµ‹è¯•é›†æŒ‡æ ‡

```python
# 7. æµ‹è¯•é›†ç»“æœè¯„ä»·

# r2_score å†³å®šç³»æ•°ï¼Œåæ˜ å› å˜é‡çš„å…¨éƒ¨å˜å¼‚èƒ½é€šè¿‡å›å½’å…³ç³»è¢«è‡ªå˜é‡è§£é‡Šçš„æ¯”ä¾‹ã€‚
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2, mse
```
